```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Leyendo los datos del archivo
df = pd.read_csv("train.csv")
x = df['GrLivArea']
y = df['SalePrice']
plt.scatter(x, y, alpha=0.5)
plt.show()
```

```python
#Obteniendo la pendiente (m) y la ordenada al origen (b)
N = len(x)
m = (np.sum(x*y) - 1/N * np.sum(x) * np.sum(y))/(np.sum(x**2)- 1/N * np.sum(x)**2)
b = (np.sum(y) - m * np.sum(x))/(N)
print("pendiente: ", m, "\nordenada: ", b)

#Obteniendo la gráfica de la linea recta
x1 = np.linspace(min(x), max(x), 1000)
y1 = m*x1 + b
```

```python
#Mostrando datos con el ajuste por mínimos cuadrados
plt.plot(x1, y1, label="Ajuste", c="b")
plt.scatter(x,y, label="Datos", c="g", alpha=.5)
plt.title("Ajuste por mínimos cuadrados")
plt.legend()
plt.show()
```

Las variables del dataframe relevantes son:
+ GrLivArea (Above ground living area) se refiere al espacio habitable en una vivienda que está por encima del nivel del suelo. En otras palabras, se trata del espacio de vida que no incluye sótanos ni áreas subterráneas.
+ SalePrice (Precio de Venta) se refiere al monto monetario por el cual una propiedad, producto o servicio es vendido.

Una relación lineal entre "GrLivArea" y "SalePrice" significaría que a medida que el tamaño del área habitable aumenta, el precio de venta también tiende a aumentar en una forma predecible.
